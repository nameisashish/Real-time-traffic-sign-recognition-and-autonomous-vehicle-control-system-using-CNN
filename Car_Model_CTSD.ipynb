```python
import os
import cv2
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense
from keras.optimizers import Adam
from keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
from keras.callbacks import Callback

# Parameters
target_size = (32, 32)
batch_size_val = 50
epochs_val = 35
steps_per_epoch_val = 2000

# File paths
file_path = '.../archive/annotations.csv'
image_dir = '.../archive/images'

# Load annotations
annotations_df = pd.read_csv(file_path)

# Process images with fallback for invalid bounding boxes
def load_and_preprocess_images_with_boxes(dataframe, image_dir, target_size=(32, 32)):
    images, labels = [], []
    skipped_images = 0
    total_images = len(dataframe)
    
    for _, row in dataframe.iterrows():
        img_path = os.path.join(image_dir, row['file_name'])
        if os.path.exists(img_path):
            img = cv2.imread(img_path)
            if img is not None:
                x1, y1, x2, y2 = row['x1'], row['y1'], row['x2'], row['y2']
                if x1 < 0 or y1 < 0 or x2 > img.shape[1] or y2 > img.shape[0] or x1 >= x2 or y1 >= y2:
                    cropped_img = img
                else:
                    cropped_img = img[y1:y2, x1:x2]
                resized_img = cv2.resize(cropped_img, target_size)
                images.append(resized_img)
                labels.append(row['category'])
            else:
                skipped_images += 1
        else:
            skipped_images += 1
    
    print(f"Total images: {total_images}")
    print(f"Images processed: {len(images)}")
    print(f"Images skipped: {skipped_images}")
    return np.array(images), np.array(labels)

# Load images
images, labels = load_and_preprocess_images_with_boxes(annotations_df, image_dir, target_size)

# Split dataset into training, validation, and testing
X_train, X_temp, y_train, y_temp = train_test_split(images, labels, test_size=0.3, random_state=42)
X_validation, X_test, y_validation, y_test = train_test_split(X_temp, y_temp, test_size=1/3, random_state=42)

# Preprocessing functions
def grayscale(img):
    return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

def equalize(img):
    return cv2.equalizeHist(img)

def preprocessing(img):
    img = grayscale(img)  # Convert to grayscale
    img = equalize(img)  # Standardize lighting
    img = img / 255.0  # Normalize to [0, 1]
    return img

# Apply preprocessing
X_train = np.array([preprocessing(img) for img in X_train])
X_validation = np.array([preprocessing(img) for img in X_validation])
X_test = np.array([preprocessing(img) for img in X_test])

# Display preprocessed image samples
plt.figure(figsize=(10, 5))
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(X_train[i], cmap="gray")
    plt.axis("off")
plt.suptitle("Sample Preprocessed Images", fontsize=14)
plt.show()

# Reshaping for CNN
X_train = X_train.reshape(X_train.shape[0], target_size[0], target_size[1], 1)
X_validation = X_validation.reshape(X_validation.shape[0], target_size[0], target_size[1], 1)
X_test = X_test.reshape(X_test.shape[0], target_size[0], target_size[1], 1)

# Augment images
dataGen = ImageDataGenerator(width_shift_range=0.1, height_shift_range=0.1, zoom_range=0.2, shear_range=0.1, rotation_range=10)
dataGen.fit(X_train)

# One-hot encode labels
num_classes = len(np.unique(labels))
y_train = to_categorical(y_train, num_classes)
y_validation = to_categorical(y_validation, num_classes)
y_test = to_categorical(y_test, num_classes)

# Define CNN model
def myModel():
    no_of_filters = 60
    filter_size1 = (5, 5)
    filter_size2 = (3, 3)
    pool_size = (2, 2)
    nodes_hidden_layer = 500

    model = Sequential([
        Conv2D(no_of_filters, filter_size1, input_shape=(target_size[0], target_size[1], 1), activation='relu'),
        Conv2D(no_of_filters, filter_size1, activation='relu'),
        MaxPooling2D(pool_size=pool_size),
        Conv2D(no_of_filters // 2, filter_size2, activation='relu'),
        Conv2D(no_of_filters // 2, filter_size2, activation='relu'),
        MaxPooling2D(pool_size=pool_size),
        Dropout(0.5),
        Flatten(),
        Dense(nodes_hidden_layer, activation='relu'),
        Dropout(0.5),
        Dense(num_classes, activation='softmax')
    ])
    model.compile(Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])
    return model

model = myModel()
print(model.summary())

# Custom callback for test loss and accuracy
class TestMetricsCallback(Callback):
    def __init__(self, test_data):
        self.test_data = test_data
        self.test_losses = []
        self.test_accuracies = []

    def on_epoch_end(self, epoch, logs=None):
        X_test, y_test = self.test_data
        test_loss, test_accuracy = self.model.evaluate(X_test, y_test, verbose=0)
        self.test_losses.append(test_loss)
        self.test_accuracies.append(test_accuracy)
        print(f"Epoch {epoch+1}: Test Loss = {test_loss:.4f}, Test Accuracy = {test_accuracy:.4f}")

test_metrics_callback = TestMetricsCallback((X_test, y_test))

# Train the model
history = model.fit(
    dataGen.flow(X_train, y_train, batch_size=batch_size_val),
    steps_per_epoch=steps_per_epoch_val,
    validation_data=(X_validation, y_validation),
    epochs=epochs_val,
    callbacks=[test_metrics_callback],
    shuffle=True
)

# Plot training and validation accuracy and loss
plt.figure(figsize=(12, 5))

# Loss
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.plot(test_metrics_callback.test_losses, label='Test Loss', linestyle='--', color='red')
plt.legend()
plt.title('Loss Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')

# Accuracy
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.plot(test_metrics_callback.test_accuracies, label='Test Accuracy', linestyle='--', color='red')
plt.legend()
plt.title('Accuracy Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')

plt.tight_layout()
plt.show()

# Confusion matrix
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = np.argmax(y_test, axis=1)

conf_matrix = confusion_matrix(y_true, y_pred_classes)
plt.figure(figsize=(10, 10))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()

# Classification report
class_report = classification_report(y_true, y_pred_classes)
print('Classification Report:')
print(class_report)
```

```python
score = model.evaluate(X_test, y_test, verbose=0)
print('Test Score:', score[0])
print('Test Accuracy:', score[1])
```
